```python
import pandas as pd

# O pandas lê o arquivo que você acabou de subir
df = pd.read_csv('SDW2023.csv')

# Exibe a tabela para você conferir se deu certo
print("Tabela carregada com sucesso:")
print(df)

# Extrai apenas os IDs para uma lista, conforme o desafio pede
user_ids = df['UserID'].tolist()
print(f"\nLista de IDs extraída para o pipeline: {user_ids}")
```

```python
# 1. Configuração da "IA" (Simulação ou Real)
def generate_ai_news(user_name):
    # Aqui criamos o "Prompt": a instrução para a IA
    prompt = f"Olá {user_name}, como estão seus investimentos? Sabia que diversificar é a chave do sucesso?"
    return prompt

# 2. Aplicando a transformação para cada usuário
# Vamos criar uma nova coluna no nosso DataFrame chamada 'News'
df['News'] = [generate_ai_news(nome) for nome in df['Nome']]

print("Transformação concluída!")
print(df[['Nome', 'News']]) # Mostra apenas o nome e a mensagem gerada
```

```python
# Salvando o resultado final em um novo arquivo
df.to_csv('SDW2023_com_mensagens.csv', index=False)

print("Arquivo 'SDW2023_com_mensagens.csv' gerado com sucesso!")
```
